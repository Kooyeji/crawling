```python
#범례 표시 
data_dict={'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
plt.plot('data_x','data_y',data=data_dict, label='Price(&)')
plt.xlabel('x축',fontdict={'color':'b','size':'x-large',}) #앞에 세줄을 씀으로 인해 x축 y축도 잘 표기된다.
plt.ylabel('y축',fontdict={'color':'r','size':'12'})
plt.legend() #범례 표시안하면 안나옴
plt.show()
```

```python
#범례 위치
data_dict={'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
plt.plot('data_x','data_y',data=data_dict, label='Price(&)')
plt.xlabel('x축',fontdict={'color':'b','size':'x-large',}) #앞에 세줄을 씀으로 인해 x축 y축도 잘 표기된다.
plt.ylabel('y축',fontdict={'color':'r','size':'12'})
plt.legend(loc=3) #1이면 우상, 2이면 좌상 3이면 좌하 10이면 정가운데
plt.show()
```

```python
#그림을 여러개 그릴 필요가 있다.
data_1={'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
data_2={'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y',data=data_1, label='Price(&)')
plt.plot('data_x','data_y',data=data_2, label='Demand(#)')

plt.xlabel('x축',fontdict={'color':'b','size':'x-large',}) #앞에 세줄을 씀으로 인해 x축 y축도 잘 표기된다.
plt.ylabel('y축',fontdict={'color':'r','size':'12'})
#plt.legend() #범례 표시는 1열이 기본
#plt.legend(ncol=2) #이러면 열이 2개 ncol옵션을 사용하면 된다. 
plt.legend(fontsize=12,shadow=True) #조금더 입체적으로 표시(shadow를 통해)
plt.show()
```

```python
#특별한 설정이 없으면 실선으로 그래프를 그리게 된다. 
#옵션으로 가능함
plt.plot([1,2,3,4],[2,3,5,10],'bo-') #선그리고 원도 그리고
plt.axis('square') #이걸 써야 이상한게 사라짐
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
#세개의 선을 
# 빨강 실선 [1,2,3,4],[2,2.5,3.3,4.3]   리플 
# 그린 점선 [1,2,3,4],[2,3,3,6]     이더리움
# 블루 삼각형 [1,2,3,4],[2,3,5,10]  비트코인
#세개 개수가 맞아야함
#범례를 표시하세요
plt.plot([1,2,3,4],[2,2.5,3.3,4.3],'r-',label='리플')
plt.plot([1,2,3,4],[2,3,3,6],'g:',label='이더리움')
plt.plot([1,2,3,4],[2,3,5,10],'b^',label='비트코인')
plt.legend(ncol=3)
plt.show()
```

```python
#난수를 발생시켜보자
import numpy as np #numpy 계산관련 라이브러리 numpy랑 pandas가 완전 기본
a=np.arange(0,2,0.2) #0부터 2미만까지 0.2씩 리스트 생성
plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo',markersize=10) #마커의 크기 설정
plt.plot(a,a**3,'g-.',linewidth=10) #선의 두꼐 설정
plt.grid(True, axis='y') #그리드 설정, y축만 설정
plt.title('그래프 제목',loc='right',fontdict={'fontsize':16,'fontweight':'bold'})
plt.show()
```

```python
#막대 그래프 bar
z=np.arange(3) #0,1,2
x=['2020','2021','2022']
y=[100,400,900]
plt.bar(z,y,width=0.6,align='edge',color='green') #가운데에 정렬되어있던게 edge가 되면 옆으로 빠진다. 
plt.xticks(z,x)
plt.show()
```

```python
#윤석열 후보, 허경영 후보를 비교할 때는 시간의 흐름보다 수평으로 나타내는게 
#더 유익
z=np.arange(3) #0,1,2
x=['2020','2021','2022']
y=[100,400,900]
plt.barh(z,y,color='green')
plt.yticks(z,x)
plt.show()
```

```python
#산점도 scatter plot 

np.random.seed(1) #난수 발생시 일정한 값으로 난수 발생 항상 같은 결과값이 나오도록
n=50
x=np.random.rand(n) #난수를 발생시키는데 0~1사이의 난수 발생
y=np.random.rand(n)
plt.scatter(x,y)
plt.show()
```

```python
np.random.seed(1) #난수 발생시 일정한 값으로 난수 발생 항상 같은 결과값이 나오도록
n=50
x=np.random.rand(n) #난수를 발생시키는데 0~1사이의 난수 발생 (%)쓸데 편함
y=np.random.rand(n)
size=(30*np.random.rand(n)) ** 2
color=np.random.rand(n) 
plt.scatter(x,y,s=size,c=color,alpha=0.4) #실행시킬 때마다, 위치도 크기도 모두 일정 
#alpha는 투명도를 조절/투명도를 조절하면 원색보다 고급지다. 
plt.show()
```

```python
#어떤색이 크다, 어떤색이 작다
np.random.seed(1) #난수 발생시 일정한 값으로 난수 발생 항상 같은 결과값이 나오도록
n=50
x=np.random.rand(n) #난수를 발생시키는데 0~1사이의 난수 발생 (%)쓸데 편함
y=np.random.rand(n)
size=(30*np.random.rand(n)) ** 2
color=np.random.rand(n) 
plt.scatter(x,y,s=size,c=color,alpha=0.4,cmap='Set1')
plt.colorbar() #색깔별로 크기까지 표현
plt.show()
```
